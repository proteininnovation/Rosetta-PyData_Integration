```python
import pandas
```

```python
import numpy
```

```python
import pylab
import seaborn
```

```python
import matplotlib
```

```python
import json
```

```python
samples = pandas.DataFrame.from_records(map(json.loads, open("profile_result.txt").readlines()))
```

```python
samples["protocol"][samples.protocol == "threaded_parallel"] = "1 node, 24 thread/worker, 24 cores"
samples["protocol"][samples.protocol == "distributed_parallel"] = "4 node, 6 thread/worker, 96 cores"
```

```python
font = {'family' : 'normal',
        'size'   : 16}

figure = {'dpi' : 300}

matplotlib.rc('font', **font)
matplotlib.rc('figure', **figure)
```

```python
pylab.grid(True)
pylab.xscale("log", basex=2)
pylab.yscale("log", basey=2)
pylab.axvline(x=24, lw=5, alpha=.3, color="blue")
pylab.axvline(x=96, lw=5, alpha=.3, color="orange")
seaborn.lineplot(x="nstruct", y="walltime", hue="protocol", markers=True, data=samples, err_style=None )
seaborn.scatterplot(x="nstruct", y="walltime", hue="protocol", data=samples, legend=False)
pylab.xlabel("Relax Batch Size")
pylab.ylabel("Batch Walltime (Seconds)")
```
